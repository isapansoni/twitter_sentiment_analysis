```python
from textblob import TextBlob
import tweepy
import pandas
```

```python
consumer_key = ''
consumer_secret= ''

access_token = ''
access_token_secret = ''
```

```python
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
```

```python
api = tweepy.API(auth)
```

```python
public_tweets = api.search('rape',count=200)
```

```python
type(public_tweets)
```

```python

tweet_list = []
polarity_list=[]
subjectivity_list= [] 

for tweets in public_tweets:
    tweet_list.append(tweets.text)
    analysis = TextBlob(tweets.text)
    polarity_list.append(analysis.sentiment.polarity)
    subjectivity_list.append(analysis.sentiment.subjectivity)
```

```python
#multiple list into dataframe
df = pandas.DataFrame({'tweets':tweet_list,                   
                       'polarity':polarity_list,
                       'subjectivity':subjectivity_list})

df.to_csv('output.csv')
```
